**Gender**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Diwali Sales Data.csv", encoding="latin1")
Gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(Gender_counts.index, Gender_counts.values, color=['Darkblue', 'Lightblue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Buyers by Gender')

for i, count in enumerate(Gender_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

```python
sales_gen = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(sales_gen.index, sales_gen.values, color=['Darkgreen','lightgreen'])
plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.title('Total Purchase Amount by Gender')

# Add labels on top of bars
for i, amount in enumerate(sales_gen.values):
    plt.text(i, amount, f'{amount:.0f}', ha='center', va='bottom')

plt.show()
```

**Age**

```python
import matplotlib.pyplot as plt

age_gender_counts = df.groupby(['Age Group', 'Gender']).size().unstack()

age_gender_counts.plot(kind='bar', figsize=(10,5))
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Buyers by Age Group and Gender')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

**State**

```python
top_orders = df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
plt.bar(top_orders.index, top_orders.values, color='blue')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Top 10 States by Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_amount = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
plt.bar(top_amount.index, top_amount.values, color='orchid')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.title('Top 10 States by Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Marital Status**

```python
marital_gender_sales = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().unstack()

marital_gender_sales.plot(kind='bar', figsize=(10, 5))
plt.xlabel('Marital Status')
plt.ylabel('Total Purchase Amount')
plt.title('Total Sales by Marital Status and Gender')
plt.legend(title='Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Occupation**

```python
import matplotlib.pyplot as plt

occupation_counts = df['Occupation'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(occupation_counts.index, occupation_counts.values, color='skyblue')
plt.xlabel('Occupation')
plt.ylabel('Number of Buyers')
plt.title('Buyers by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
occupation_sales = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(occupation_sales.index, occupation_sales.values, color='orange')
plt.xlabel('Occupation')
plt.ylabel('Total Amount Spent')
plt.title('Total Purchase Amount by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Product Category**

```python
product_counts = df['Product_Category'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(product_counts.index, product_counts.values, color='plum')
plt.xlabel('Product Category')
plt.ylabel('Count of Products Sold')
plt.title('Products Sold by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df_cleaned = df.dropna(subset=['Product_Category'])
category_sales = df_cleaned.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
bars = plt.bar(category_sales.index, category_sales.values, color=plt.cm.tab20.colors)
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales by Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show() 
```
